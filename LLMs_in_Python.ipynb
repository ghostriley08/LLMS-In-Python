```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="t5-small")
text = (
    "Agentic AI systems represent a significant evolution from traditional "
    "large language models (LLMs). While standard LLMs act as passive "
    "query-response engines—waiting for a user prompt, generating text, "
    "and then stopping—Agentic AI involves systems capable of autonomous "
    "planning, tool utilization, and iterative problem-solving. At the "
    "core of these systems are state-management frameworks that allow "
    "developers to construct complex workflows where multiple AI agents "
    "can collaborate, delegate, or even debate."
)
summary = summarizer(text,max_length = 50,clean_up_tokenization_spaces=True)
print(summary[0]['summary_text'])
```

```python
# Using Pre-Trained LLMS
# Text Generation
from transformers import pipeline
text = ("I recently stayed at the Riverview Hotel and it was a wonderful experience."
"The room was clean, the staff was incredibly friendly, and the view of the river from our balcony was stunning." # Customer Review
"The breakfast buffet was also delicious. We will definitely be coming back for our next vacation!")

generator = pipeline("text-generation",model = "gpt2") # The text generator
```

```python

response = "Dear valued customer, I am glad to hear you had a good stay with us." # Respone
prompt = f"Customer review:\n{text}\n\nHotel reponse to the customer:\n{response}" # Prompt
output = generator(prompt,max_length=200,clean_up_tokenization_spaces=True,pad_token_id = generator.tokenizer.eos_token_id)
print(output[0]['generated_text'])
```

```python
# Translation
spanish_text = "Este curso sobre LLMs se está poniendo muy interesante"

translator = pipeline("translation_es_to_en",model = "Helsinki-NLP/opus-mt-es-en")
translations = translator(spanish_text,clean_up_tokenization_spaces = True)
print(translations[0]['translation_text'])
```
