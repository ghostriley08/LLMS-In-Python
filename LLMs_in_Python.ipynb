```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="t5-small")
text = (
    "Agentic AI systems represent a significant evolution from traditional "
    "large language models (LLMs). While standard LLMs act as passive "
    "query-response engines—waiting for a user prompt, generating text, "
    "and then stopping—Agentic AI involves systems capable of autonomous "
    "planning, tool utilization, and iterative problem-solving. At the "
    "core of these systems are state-management frameworks that allow "
    "developers to construct complex workflows where multiple AI agents "
    "can collaborate, delegate, or even debate."
)
summary = summarizer(text,max_length = 50,clean_up_tokenization_spaces=True)
print(summary[0]['summary_text'])
```

```python
# Using Pre-Trained LLMS
# Text Generation
from transformers import pipeline
text = ("I recently stayed at the Riverview Hotel and it was a wonderful experience."
"The room was clean, the staff was incredibly friendly, and the view of the river from our balcony was stunning." # Customer Review
"The breakfast buffet was also delicious. We will definitely be coming back for our next vacation!")

generator = pipeline("text-generation",model = "gpt2") # The text generator
```

```python

response = "Dear valued customer, I am glad to hear you had a good stay with us." # Respone
prompt = f"Customer review:\n{text}\n\nHotel reponse to the customer:\n{response}" # Prompt
output = generator(prompt,max_length=200,clean_up_tokenization_spaces=True,pad_token_id = generator.tokenizer.eos_token_id)
print(output[0]['generated_text'])
```

```python
# Translation
spanish_text = "Este curso sobre LLMs se está poniendo muy interesante"

translator = pipeline("translation_es_to_en",model = "Helsinki-NLP/opus-mt-es-en")
translations = translator(spanish_text,clean_up_tokenization_spaces = True)
print(translations[0]['translation_text'])
```

```python
# Transformers
# Encoder Only
llm = pipeline(model = "bert-base-uncased")
print(llm.model)
# Focuses on ouput
```

```python
print(llm.model.config.is_decoder)
print(llm.model.config.is_encoder_decoder)
```

```python
# Identifying a Decoder
llm = pipeline(model="gpt2")
print(llm.model.config)
# Text generation in the ouptut indicates that it is a decoder
```

```python
# Encoder - Decoder
# Found in T5 , BART models
# Helps both in input and ouptut processing
llm = pipeline("translation_es_to_en", model="Helsinki-NLP/opus-mt-es-en")

print(llm.model)
print(llm.model.config.is_encoder_decoder)
```

```python
# Extractive - Generative Question answering
# Extractive answering requires context , encoder only models, within context.
# Generative answering can go out of context , requires decoder only or encoder - decoder models .
text = (
    "The Mona Lisa is a half-length portrait painting created by the Italian "
    "Renaissance master Pablo Picasso. It is widely considered the most famous "
    "painting in the world. Believed to have been painted between 1503 and 1506, "
    "it features Lisa Gherardini, the wife of a wealthy Florentine merchant named "
    "Francesco del Giocondo. The painting is renowned for its subject's enigmatic "
    "expression, monumental composition, and subtle modeling of forms. It was "
    "acquired by King Francis I of France and is now the property of the French Republic. "
    "Today, it is permanently displayed behind bulletproof glass at the Louvre Museum "
    "in Paris, where it attracts millions of visitors annually."
) # Context
question = "Who painted the Mona Lisa?"

# Define the appropriate model
qa = pipeline(task="question-answering", model="distilbert-base-uncased-distilled-squad") # Encoder only model

output = qa(question=question, context=text)
print(output['answer'])
# It did not go out of context , although the actual answer to this is Mona Lisa.
```

```python
text = (
    "The Mona Lisa is a half-length portrait painting created by the Italian "
    "Renaissance master Pablo Picasso. It is widely considered the most famous "
    "painting in the world. Believed to have been painted between 1503 and 1506, "
    "it features Lisa Gherardini, the wife of a wealthy Florentine merchant named "
    "Francesco del Giocondo. The painting is renowned for its subject's enigmatic "
    "expression, monumental composition, and subtle modeling of forms. It was "
    "acquired by King Francis I of France and is now the property of the French Republic. "
    "Today, it is permanently displayed behind bulletproof glass at the Louvre Museum "
    "in Paris, where it attracts millions of visitors annually."
) # Context
question = "Who painted the Mona Lisa?"

# Define the appropriate model
generator = pipeline(task="text-generation", model="gpt2")
# Properly formatted text
input_text = f"Context: {text}\n\nQuestion: {question}\n\nAnswer:"
# 3. Generate the text
output = generator(input_text, max_new_tokens=20)
print(output[0]['generated_text'])
# Output shows out of context knowledge , 1900s was nowhere in the context
```
